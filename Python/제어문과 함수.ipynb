## for
- 파이썬에서는 매우 유용하게 사용되는 문법
  - 시퀀스 타입에 특화 되어 있습니다. 
  - 리스트, 튜플, 딕셔너리, ... 

```
for 변수 in 이터레이블:
```

````python
# 딱 원소의 갯수만큼 동작
# 반복하면서 시퀀스의 원소를 첫 번째 원소부터 순차적으로, 매 회차마다 변수로 입력 받습니다. 
for x in [10, 20, 30]:
  print(x)
````

range와 함께 사용되는 경우 
- range도 시퀀스 타입
  - 범위 내의 수열을 생성

````python
# 0부터 9까지의 수열
# 마지막 원소를 포함하지 않습니다. 
list( range(10) )
````

````python
# 시작 범위가 주어지는 경우
# 5부터 9까지의 수열
list( range(5, 10) )
````

````python
# 슬라이스와 마찬가지로 step을 정할 수 있습니다. 
# 홀수 수열, 짝수 수열을 만들수 있습니다. 
print( list( range(1, 10, 2) ) )
print( list( range(0, 10, 2) ) )
````

````python
# for i in list( range(10) ):
for i in range(10):
  print(i)
````

list comprehension
- 수하적으로 집합을 정의하는 경우
$$
  A = \left\{x | x \le 10 \right\}
$$

- 파이썬에서는 `|` 를 제외하고, 원소 x에 대한 정의를 for loop를 이용

````python
A = [ x for x in range(11) ]
A
````

````python
# 혹은 초기화된 리스트가 필요한 경우
# 원소는 필요가 없고, 원소의 갯수만큼만 반복을 하는 경우 
# 변수 이름 대신에 '_'를 사용해서 표현
[0 for _ in range(10)]
````

# 함수
- 각 언어마다 표현이 다른거지 함수의 동작은 동일합니다. 

````python
def func(a, b):
  return a + b
````

````python
# 입력값은 정의된 함수의 파라미터의 갯수와 순서를 맞춰서 넣어주면 됩니다. 
func(10, 20)
````

````python
# 파이썬은 입력 파라미터를 직접 지정하는 경우에는 순서를 맞추지 않아도 됩니다. 
func(b = 10, a = 20)
````

디폴트 파라미터에 대한 정의

````python
def func(a=0, b=0):
  return a + b
````

````python
func()
````

````python
# 디폴트 값이 있는 파라미터가 없는 파라미터 앞에 올 수 없습니다. 
def func(a = 0, b):
  return a + b
````

````python
# 이렇게 되면 인터프리터는 입력값을 a에 전달해야 되는지 b에 전달해야되는지 알 수 없습니다. 
func(10)
````

````python
# 가변인자
# packing 이라고 표현 합니다. 
def func( *args ):
  print(args)
````

````python
# 정확하게는 여러개의 입력값을 하나의 튜플로 묶어서 전달(Packing)
func(1, 2, 3, 4, 5, 6, 7, 8, 9, 0, )
````

````python
def func( **kwargs ):
  print(kwargs)
````

````python
# 딕셔너리로 형태로 묶어서 전달
func( a = 10, b = 20, c = 30)
````

````python
# 반대인 경우도 가능합니다.(Unpacking)
def func(a, b, c):
  print(a, b, c)
````

````python
func( *[1, 2, 3] )
````

````python
# 이런걸 응용하면 이렇게도 사용이 가능합니다. 
a, b = [1, 2]
print(a, b)
````

````python
# 더 응용하면 이렇게 사용할 수 있습니다. 
a, b, *c = [1, 2, 3, 4, 5, 6, 7, ]
print(a, b, c)
````

## 익명함수(람다)
- 함수의 이름이 없기 때문에 익명함수라고 부릅니다. 
  - 간단하게 한 번 사용되고 마는, 그런 함수를 정의할 때 사용

````python
(lambda a, b: a + b)(10, 20)
````

## 함수형 문법
- 절차적 프로그래밍
- 객체지향 프로그래밍
- 함수형 프로그래밍
  - map, filter, ... 
  - 집합과 함수가 정의되면 반복문 없이 y를 정의

````python
# 예를 들어서 실수집합 전체에 대해서 정수로 변환하고 싶다고 가정
# x = [실수집합]
# y = [정수집합]
# int()

x = [1., 2., 3., 4.]
y = [] 
for i in x:
  y.append( int(i) )
````

````python
y
````

````python
# 함수형 문법을 사용하는 경우
list( map(int, x) )
````
