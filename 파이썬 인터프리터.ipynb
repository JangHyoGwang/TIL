- 컴파일 과정이 없습니다. 
- REPL(Read, Evaluation, Print, Loop)
  - Read: 입력된 명령어를 읽고
    - 중간에서 해석(문자 -> 숫자)
  - Evaluation: 명령어를 실행

  - Loop: 이 과정을 계속 반복
- 주피터는 일반적인 파이썬의 실행환경과 인터프리터의 실행환경의 중간적인 환경을 제공

split과 같은 경우는 `화이트 스페이스`를 기준으로 문자열을 분할

````python
'Lorem ipsum dolor sit amet'.split()
````

````python
# 어떤 파라미터의 입력인지 지정해도 됩니다.
# 파라미터의 순서를 고려하지 않아도 됩니다. 
'Lorem/ipsum/dolor/sit/amet'.split(maxsplit=10, sep='/')
````

## 얕은 복사(주소 복사) Vs. 깊은 복사(값 복사)

````python
arr
````

````python
# 새로운 메모리로 복사
other = arr
````

````python
print( hex(id(other)))
print( hex(id(arr)))
````

````python
# copy를 이용해서 새로운 메모리에 복사
other = arr.copy()
````

````python
print( hex(id(other)))
print( hex(id(arr)))
````

````python
# 슬라이스를 이용해도 동일한 결과를 얻을 수 있습니다. 
other = arr[:]
````

````python
print( hex(id(other)))
print( hex(id(arr)))
````

그런데 왜 `얕은복사`라는 표현을 사용할까? 
- 분명히 값이 복사된게 맞는데
- 완벽하게 깊은 복사가 되지는 않습니다. 

````python
# 파이썬에서의 다차원 리스트는 리스트를 리스트를 원소로 갖는 형태
arr2D = [
  [1, 2, 3], [4, 5, 6], [7, 8, 9]
]
````

````python
arr2D
````

````python
# 파이썬의 리스트는 타입이 모두 동일할 필요가 없습니다. 
arr = [1, 'string', [1, 2, 3]]
arr
````

````python
arr2D[0][0]
````

````python
other = arr2D[:]
````

````python
print( hex(id(other)))
print( hex(id(arr2D)))
print( hex(id(other[0])))
print( hex(id(arr2D[0])))
````

파이썬에서의 깊은 복사는 어떻게 할 수 있는가? 
- deepcopy를 이용하면 완벽하게 깊은복사를 할 수 있습니다. 
- 그렇지 않으면, 값의 일부는 복사가 되고, 일부는 복사가 안될 수 있습니다. 

````python
import copy
other = copy.deepcopy(arr2D)
````

````python
print( hex(id(other)))
print( hex(id(arr2D)))
print( hex(id(other[0])))
print( hex(id(arr2D[0])))
````

- set과 동일한 기호를 사용
  - 자료의 형태가 타입을 결정

```
  {key: value, key:value, ... }
```

````python
# 빈 닉셔너리를 생성하는 경우
# 원소가 아무것도 없으면 딕셔너리로 판단
hash = {}
type(hash)
````

````python
hash = dict()
type(hash)
````

````python
# 키가 존재하면 value를 업데이트 하고, 키가 존재하지 않으면, 새로운 키를 만들게 됩니다. 
hash[0] = 10
hash
````

````python
hash['string'] = 20
hash
````

````python
# 키와 값을 따로 확인
print( hash.keys() )
print( hash.values() )
print( hash.items() )
````

````python
for key in hash:
  print(key)
````

````python
for value in hash.values():
  print(value)
````

````python
for key, value in hash.items():
  print(key, value)
````
